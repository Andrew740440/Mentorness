```python
# find the avg percentage of active workers engaged to the number of workers registered under NREGA per state
```

```python
df.head()
```

```python
df_state_ratio = df[['state_name','Total No. of Workers','Total No. of Active Workers']]
```

```python
df_state_ratio['Registered_To_Engaged_Ratio'] = (df_state_ratio['Total No. of Active Workers']/df_state_ratio['Total No. of Workers']) * 100
```

```python
df_state_ratio.head()
```

```python
df_state_ratio = df_state_ratio.groupby(['state_name'],as_index = True).mean()
```

```python
df_state_ratio.sort_values(by = 'Registered_To_Engaged_Ratio',ascending = False,inplace =True)
```

```python
df_state_ratio['Registered_To_Engaged_Ratio'].plot(kind = 'bar',figsize = (15,5))
plt.xlabel('State')
plt.ylabel('percentage')
plt.title('Percent distribution of Engaged Active Workers to Registered Worker per State')
plt.show()
```

```python
# find if there is a corrolation between percentage of engaged workers to the approved budget per state
```

```python
df_corr = df[['state_name','Total No. of Workers','Total No. of Active Workers','Approved Labour Budget']]
```

```python
df_corr['Registered_for_Engagement'] = (df_corr['Total No. of Active Workers']/df_corr['Total No. of Workers']) * 100
```

```python
df_corr.head()
```

```python
df_corr = df_corr.groupby(['state_name'],as_index = True).sum()
```

```python
df_corr.head()
```

```python
df_corr[['Approved Labour Budget','Registered_for_Engagement']].corr()
```

```python
# There is a positive corrolation between the Approved Labour Budget and the percentage of persons employed per state
```

```python
import seaborn as sns
```

```python
sns.regplot(x = 'Registered_for_Engagement',y = 'Approved Labour Budget',data = df_corr)
plt.title('Corrolation plot of Registered_for_Engagement against Approved_Labour_Budget')

plt.show()
```
